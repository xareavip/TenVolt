```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xPoints=np.array([0,6])
yPoints=np.array([0,100])

plt.plot(xPoints,yPoints,'mo')
plt.show()
```

```python
xPoints=np.array([1,3,6,8])
yPoints=np.array([3,8,1,10])
z=np.array([2,8,3,4])
plt.subplot(2,2,1)
plt.plot(xPoints,yPoints)


plt.subplot(2,2,2)
plt.plot(xPoints,yPoints,marker='.')

plt.subplot(2,2,3)
plt.plot(xPoints,yPoints,'o-r',ms=20,mec='green',mfc='yellow',linewidth=2)
plt.plot(xPoints,z)
plt.grid(axis='x',color='r',linestyle='--',linewidth=0.5)
plt.title('Scatter Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')


plt.subplot(2,2,4)
explode = (0, 0.1, 0, 0)
plt.pie(xPoints,labels=['A','B','C','D'],shadow=True,explode=explode)
plt.show()
```
